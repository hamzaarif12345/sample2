```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import required module from sklearn (scikit library)
from sklearn.linear_model import LinearRegression
```

```python
# import sample dataset generator
from sklearn.datasets import make_regression
```

```python
# develop sample observations (by default linearly separable)
X,Y = make_regression(n_features=1, noise=5, n_samples=100)
```

```python
# lets plot scatter plot diagram
plt.xlabel('Independent Variable - X')
plt.ylabel('Dependent Variable - Y')
plt.scatter(X,Y,s=4)
```

```python
lr = LinearRegression()
```

```python
# now fit a line to the observations/sample data
lr.fit(X,Y)
```

```python
# b0 and b1 are found
lr.coef_, lr.intercept_
```

```python
# predict by using 'lr' model developed above
pred = lr.predict(X)
```

```python
# lets draw line closely fitting to data points/observations/sample dataset
plt.scatter(X,Y,s=3, label='training')
plt.scatter(X,pred,s=3, label='prediction')
plt.xlabel('Independent Variable - X')
plt.ylabel('Dependent Variable - Y')
plt.legend()
plt.show()
```
