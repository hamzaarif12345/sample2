```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

### Plots, Axes, labels, legends


```python
#creating the values between the 1 to 10 with the gap of 20
x_axis_values = np.linspace(1, 10, 20)
x_axis_values
```

```python
y_axis_values = np.log(x_axis_values)
y_axis_values
```

```python
# Line plot

plt.plot(x_axis_values, y_axis_values, 'blue')
plt.xlabel('X Axis Values')
plt.ylabel('Y Axis Values')
plt.title('Basic plot')
plt.show()
```

```python
# When we are running analysis, we might want to compare results.
# For that we might need to have the plots placed side by side.

# Pyplot API provides subplot as an option.
plt.subplot(121)    # plt.subplot(nrows, ncols, plot_number)
plt.plot(x_axis_values, y_axis_values, 'blue')

plt.subplot(122)
plt.plot(x_axis_values, np.power(x_axis_values,2) , 'green');
```

So whatever we were seeing was the pyplot module API.

This API is good for quick plots, but Matplotlib recommends using object-oriented API as it provides more control and customization for the plots.

Most of these functions also exist as methods in the matplotlib.axes.Axes class.

Introduction to the Object Oriented Method

The main idea in using the more formal Object Oriented method is to create figure objects and then just call methods or attributes off of that object. This approach is nicer when dealing with a canvas that has multiple plots on it.

To begin we create a figure instance. Then we can add axes to that figure:
    
So let us get started.
 

```python
fig, ax = plt.subplots()   # plt.subplots is equivalent to using fig = plt.figure() and then ax = fig.add_subplot(111)

ax.plot(x_axis_values, y_axis_values, 'blue')
ax.set_xlabel('X Axis Values') # Notice the use of set_ to begin methods
ax.set_ylabel('Y Axis Values')
ax.set_title('Set Title')
```

```python
# Creates two subplots

f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(x_axis_values, y_axis_values)
ax1.set_title('Sharing Y axis')
ax2.scatter(x_axis_values, y_axis_values)
```

```python
# There are instances when we would have multiple lines in the same plot.

fig, ax = plt.subplots()
ax.plot(x_axis_values, np.power(x_axis_values,2), label="Square")
ax.plot(x_axis_values, np.power(x_axis_values,3), label="Cube")
ax.legend()  # Legend identify the line.
```
