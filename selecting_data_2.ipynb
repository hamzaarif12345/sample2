```python
import pandas as pd
```

```python
countries_data = pd.read_csv('top_10_countries.csv')
```

### Selecting rows or columns that match a list of potential values

```python
countries_data
```

You can filter your DataFrame using isin(), rows containing values in the list will be returned

```python
# Checking the Region column to see if it contians 'Asia' or 'Americas'
countries_data['Region'].isin(['Asia','Americas'])
```

```python
# Returning only rows where Region='Asia' or Region='Americas'
countries_data[countries_data['Region'].isin(['Asia','Americas'])]
```

```python
# Returning only rows where Region is NOT'Asia' and Region is NOT'Americas'
countries_data[~countries_data['Region'].isin(['Asia','Americas'])]
```
