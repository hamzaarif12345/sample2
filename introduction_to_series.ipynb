# Series

```python
import pandas as pd
```

```python
index_list = [0,1,2,3,4,5]
```

```python
data_list = [100,200,300,400,500,600]
```

```python
# creating the series with the index_list as the index and data_list as the data
series_1 = pd.Series(data=data_list,index=index_list)
```

```python
series_1
```

```python
# You can index your Series like a python array such as a list
series_1[3]
```

#### The datatypes do not need to be the same type

```python
month_index_2021 = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
```

```python
revenue_data_2021 = (212861,100372,128953,100163,208129,182944,236284,243133,198641,158276,166398,178211)
```

```python
revenue_series = pd.Series(data=revenue_data_2021,index=month_index_2021)
```

```python
revenue_series
```

```python
# As you can see the type is Series
type(revenue_series)
```

```python
# The data does not need to be of the same type for each element
revenue_data_2021 = (212861,100372,128953,100163,208129,182944,236284,243133,198641,158276,166398,'No data provided')
revenue_series = pd.Series(data=revenue_data_2021,index=month_index_2021)
revenue_series
```

```python
revenue_data_2021 = (212861,100372,128953,100163,208129,182944,236284,243133,198641,158276,166398,178211)
revenue_series = pd.Series(data=revenue_data_2021,index=month_index_2021)
```

```python
# You can return the value for a single index by passing in the index key
revenue_series['Dec']
```

```python
# You can return the values for a multiple indexs by passing in the index keys inside of a list
revenue_series[['Jan','Mar','Apr']]
```

#### You can change the datatype

```python
# The datatype is currently int64
revenue_series
```

```python
# You can change the datatype by passing in the datatype or its character code
revenue_series = pd.Series(data=revenue_data_2021, index=month_index_2021,dtype='g')
```

```python
revenue_series
```

#### With dictionaries the keys are read as indexes and the values ar read as data

```python
dict_1={'key1':'value1','key2':'value2','key3':'value3'}
```

```python
pd.Series(data=dict_1,index=dict_1)
```

```python
pd.Series(dict_1)
```
